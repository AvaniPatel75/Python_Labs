# List

### 01) WAP to find sum of all the elements in a List.

```python
size=int(input("enter size of the list"))
li=[]
for i in range(0,size):
    li.append(int(input("Enter element of the list")))
total=sum(li)
print(f"Total of the list:{total}")
```

### 02) WAP to find largest element in a List.

### 04) WAP to interchange first and last elements in a list.

```python
size=int(input("enter size of the list"))
li=[]
for i in range(0,size):
    li.append(int(input("Enter element of the list")))
temp=li[0]
li[0]=li[len(li)-1]
li[len(li)-1]=temp
print(li)
```

### 09) WAP to count unique items in a list.

```python
size=int(input("enter size of the list"))
li=[]
count=0;
unique_items=[]
for i in range(0,size):
    li.append(int(input("Enter element of the list")))
for i in li:
    if i not in unique_items:
        unique_items.append(i)
        count+=1
print(f"Count of the unique items in list {count}")
```

### 12) WAP to count occurrences of an element in a list.

```python
size=int(input("enter size of the list"))
li=[]
count=0;
for i in range(0,size):
    li.append(int(input("Enter element of the list")))
element=int(input("enter element"))
for i in li:
    if element==i:
        count+=1
print(f"Count of the item {element} in list {count}")
```

### 14) WAP to extract elements with frequency greater than K.

```python
size=int(input("enter size of the list"))

li=[]
result=[]
for i in range(0,size):
    li.append(int(input("Enter element of the list")))
freq=int(input("enter frequency"))
for i in range(len(li)):
    count=0;
    for j in range(len(li)):
        if(li[j]==li[i]):
            count+=1;

    if count>freq and li[i] not in result:
        result.append(li[i])
print(result)
```

### 17) WAP to create a list of common elements from given two lists.

```python
size1=int(input("enter size of the list"))
list1=[]
for i in range(0,size1):
    list1.append(input("Enter number"))
size2=int(input("enter size of the list"))
list2=[]
for i in range(0,size2):
    list2.append(input("Enter number"))
ans=[]
for x in list1:
    if x in list2:
        ans.append(x)
ans
```
