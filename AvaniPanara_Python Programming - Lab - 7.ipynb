# Tuple

### 01) WAP to find sum of tuple elements.

```python
size=int(input("Enter size of the tuple"))
li=[]
for i in range(0,size):
    value=int(input("Enter element"))
    li.append(value)
t=tuple(li)
total=sum(t)
print(f"Sum of the elements of the tuple {total}")
```

### 02) WAP to find Maximum and Minimum K elements in a given tuple.

```python
size=int(input("Enter size of the tuple"))
li=[]
for i in range(0,size):
    value=int(input("Enter element"))
    li.append(value)
t=tuple(li)
print(f"Maximum element of the tuple is {max(t)}")
print(f"Maximum element of the tuple is {min(t)}")
```

### 03) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
li_size=int(input("Enter size of the list"))
li=[]
for i in range(0,li_size):
    tp_size=int(input("Enter size of the tuple"))
    temp=[]
    for j in range(0,tp_size):
        temp.append(int(input("Enter element")))
    t=tuple(temp)
    li.append(t)
print(li)
k=int(input("Enter k element"))
result=[]
for t in li:
    if all(i%k==0 for i in t):
        result.append(t)
print(result)
```

### 04) WAP to create a list of tuples from given list having number and its cube in each tuple.

```python
li_size=int(input("Enter size of the list"))
li=[]
for i in range(li_size):
    li.append(int(input("Enter number")))
print(f"Given list {li}")
result=[]
for i in li:
    cube=[i,i**3]
    result.append(tuple(cube))
print(result)    
```

### 05) WAP to find tuples with all positive elements from the given list of tuples.

```python
li_size=int(input("Enter size of the list"))
li=[]
for i in range(0,li_size):
    tp_size=int(input("Enter size of the tuple"))
    temp=[]
    for j in range(0,tp_size):
        temp.append(int(input("Enter element")))
    t=tuple(temp)
    li.append(t)
print(li)
result=[]
for tp in li:
    if all(i>0 for i in tp):
        result.append(tp)
result
```

### 07) WAP to remove tuples of length K.

```python
li_size=int(input("Enter size of the list"))
li=[]
for i in range(0,li_size):
    tp_size=int(input("Enter size of the tuple"))
    temp=[]
    for j in range(0,tp_size):
        temp.append(int(input("Enter element")))
    t=tuple(temp)
    li.append(t)
print(li)
result=[]
k=int(input("Enter number of length"))
for i in li:
    if len(i)!=k:
        result.append(i)
print(result)
```

### 08) WAP to remove duplicates from tuple.

```python
li=[]
result=[]
size=int(input("Enter size of the tuple"))
for i in range(size):
    li.append(int(input("Enter element")))
tp=tuple(li)
for i in tp:
    if i not in result:
        result.append(i)
print(tuple(result))
```

### 09) WAP to multiply adjacent elements of a tuple and print that resultant tuple.

```python
li=[]
result=[]
size=int(input("Enter size of the tuple"))
for i in range(size):
    li.append(int(input("Enter element")))
tp=tuple(li)
for i in range(len(tp)-1):
    result.append(tp[i]*tp[i+1])
print(tuple(result))
```

### 10) WAP to test if the given tuple is distinct or not.

```python
li=[]
result=[]
size=int(input("Enter size of the tuple"))
for i in range(size):
    li.append(int(input("Enter element")))
tp=tuple(li)
for i in tp:
    if i not in result:
        result.append(i)
if len(result)==len(tp):
    print(f"{tp} is distinct")
else:
    print(f"{tp} is not distinct")
```
