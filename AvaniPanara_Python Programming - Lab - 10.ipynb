### 04) WAP to find the longest word(s) in a file

```python
fp=open('first.txt',"r")
li=fp.readlines()
max=0
maxWord=""
for i in li:
    l1=i.split(" ")
    for j in l1:
        if max<=len(j):
            max=len(j)
            maxWord=j
print(maxWord)
fp.close()
```

### 05) WAP to count the no. of lines, words and characters in a given text file.

```python
char=0;
word=0
fp=open("first.txt","r")
li=fp.readlines()
print(f"No of Lines in file {len(li)}")
for i in li:
    l1=i.split()
    word+=len(l1)
    for j in l1:
        char+=len(j)
print(f"No of words in file {word}")
print(f"No of characters in file {char}")
fp.close()
```

### 08) WAP to create an UDF named frequency to count occurances of the specific word in a given text file.

```python
def FrequencyWord(filePath,word):
    count=0
    fp=open(filePath,"r")
    li=fp.readlines()
    for i in li:
        l1=i.split()
        for j in l1:
            if j==word:
                count+=1
    return count
    fp.close()
filePath="first.txt"
word=input("enter word you want to find in file:")
count=FrequencyWord(filePath,word)
print(f"frequency of the word ''{word}' is {count}")
```

### 10) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
fp=open("Primenumbers.txt","w")
flag=0
for i in range(1,100):
    for j in range(2,i):
        if i%j==0:
            flag=0
            break;
        else:
            flag=1
    if flag==1:
        fp.write(str(i)+"\n")
    else:
        pass
fp.close()
```
