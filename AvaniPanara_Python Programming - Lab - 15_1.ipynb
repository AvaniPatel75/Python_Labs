### 03) WAP to create Circle class with area and perimeter function to find area and perimeter of circle.

```python
import math
class Circle:
    def __init__(self,r):
        self.radius=r
    def areaOfCircle(self):
        return math.pi*(self.radius**2)
    def perimeterOfCircle(self):
        return 2*math.pi*self.radius
c1=Circle(5)
area=c1.areaOfCircle()
peri=c1.perimeterOfCircle()
print(area)
print(peri)
```

### 09) Define Time class with hour and minute as data member. Also define addition method to add two time objects.

```python
class Time:
    def __init__(self, hour, minute): 
        self.hour = hour
        self.minute = minute
    def addTime(self, t2):
        # Add minutes and calculate extra hours
        total_minutes = self.minute + t2.minute
        extra_hours = total_minutes // 60
        minutes = total_minutes % 60

        # Add hours including the extra hours
        hours = self.hour + t2.hour + extra_hours
        print(f"Time after addition: {hours}:{minutes:02d}")
# Example usage
t1 = Time(3, 40)
t2 = Time(4, 45)
t1.addTime(t2)
```
