### 11) WAP to demonstrate the use of date time module.

```python
import datetime as dt

# Get the current date and time
current_datetime = dt.datetime.now()
print("Current Date and Time:", current_datetime)

# Get today's date
today = dt.date.today()
print("Today's Date:", today)

# Formatting date
formatted_date = today.strftime("%d-%m-%Y")
print("Formatted Date:", formatted_date)

# Getting individual date components
print("Year:", today.year)
print("Month:", today.month)
print("Day:", today.day)

# Creating a specific date
specific_date = dt.date(2023, 5, 17)
print("Specific Date:", specific_date)

# Creating a specific time
specific_time = dt.time(14, 30, 45)
print("Specific Time:", specific_time)

# Date arithmetic: Adding days
future_date = today + dt.timedelta(days=10)
print("Date after 10 days:", future_date)

# Date arithmetic: Subtracting days
past_date = today - dt.timedelta(days=5)
print("Date 5 days ago:", past_date)

# Calculating difference between two dates
date1 = dt.date(2023, 1, 1)
date2 = dt.date(2024, 1, 1)
difference = date2 - date1

print("Difference between two dates:", difference.days, "days")
# Getting the current time
current_time = dt.datetime.now().time()
print("Current Time:", current_time)

# Formatting time
formatted_time = current_time.strftime("%H:%M:%S")
print("Formatted Time:", formatted_time)

# Getting current UTC time
utc_time = dt.datetime.utcnow()
print("Current UTC Time:", utc_time)
```

### 12) WAP to demonstrate the use of the math module.

```python
import math

num = -25.7
print("Absolute Value:", math.fabs(num))
print("Ceiling Value:", math.ceil(num))

num = 16
print("Square Root of", num, ":", math.sqrt(num))
print("2 raised to the power 3:", math.pow(2, 3))
print("Factorial of 5:", math.factorial(5))
print("GCD of 36 and 48:", math.gcd(36, 48))

angle = math.radians(30) # Convert degrees to radians
print("Sin(30 degrees):", math.sin(angle))
print("Cos(30 degrees):", math.cos(angle))

num = 100
print("Natural Log (ln) of", num, ":", math.log(num))
print("Log base 10 of", num, ":", math.log10(num))
print("Exponential (e^3):", math.exp(3))
print("Value of Pi:", math.pi)
print("Value of Euler’s Number (e):", math.e)
print("Value of Tau:", math.tau)

side_a = 3
side_b = 4
print("Hypotenuse of Triangle with sides 3 and 4:", math.hypot(side_a, side_b))
```
