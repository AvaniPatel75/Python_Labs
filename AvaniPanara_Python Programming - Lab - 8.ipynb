### 08) WAP to remove common elements of set A & B from set A.

```python
size1=int(input("Enter size of the set"))
s1=set()
for i in range(size1):
    s1.add(int(input("Enter number")))
size2=int(input("Enter size of the set"))
s2=set()
for i in range(size2):
    s2.add(int(input("Enter number")))
intersect=s1.intersection(s2)
print(f"Common elments of {s1} and {s2} is {s1.intersection(s2)}")
for i in intersect:
    if i in s1:
        s1.remove(i)
print(s1)
```
