### 01) WAP to check whether the given string is palindrome or not.

```python
str=input("Enter string")
str_rev=str[::-1]
if(str==str_rev):
    print(f"{str} is palindrom string.")
else:
    print(f"{str} is palindrom  not string.")
```

### 02) WAP to reverse the words in the given string.

```python
str=input("Enter string")
str_rev=""
li=str.split(" ")
for i in li:
    str_rev=str_rev+" "+i[::-1]
print(f"Reverese string : {str_rev}")
```

### 10) : Display credit card number. 
### card no. : 1234 5678 9012 3456
### display as  : **** **** **** 3456

```python
card_no=input("Enter card number in format XXXX XXXX XXXX XXXX")
ans = ""
count = 0 
for ch in card_no:
    if ch == " ":
        ans += " "
    else:
        count += 1
        if count <= 12:  
            ans += "*"
        else:         
            ans += ch
print(ans)
```

### 11) : Checking if the two strings are Anagram or not.
### s1 = decimal and s2 = medical are Anagram

```python
s1=input("Enter first string")
s2=input("Enter second string ")
s1=s1.lower()
s2=s2.lower()
s1=s1.replace(" ", "")
s2=s2.replace(" ", "")
list1=list(s1)
list2=list(s2)
list1.sort()
list2.sort()
if list1==list2:
    print("Anagram")
else:
    print("Not Anagram")
```

### 12) : Rearrange the given string. First lowercase then uppercase alphabets.
### input : EHlsarwiwhtwMV
### output : lsarwiwhtwEHMV

```python
s=input("Enter first string") 
lower_part=""
upper_part=""
for ch in s:
    if ch.islower():      
        lower_part+=ch
    elif ch.isupper():     
        upper_part+=ch
result=lower_part+upper_part
print(result)
```
