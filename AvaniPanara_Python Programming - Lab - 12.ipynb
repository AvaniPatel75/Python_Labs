### 01) WAP to store student names in a list. Use list methods to add names, insert a new name at a specific position, remove a name, sort the list, reverse it, and display the final result.

### 04) 4. Unique Word Analyzer
<h4>Write a Python program that reads a text paragraph from a file named article.txt.</h4><br>
• Extract all words and store them in a list.<br>
• Convert the list to a set to determine the unique words.<br>
• Create a dictionary where each key = word, and value = frequency.<br>
• Store the frequency dictionary in a new file wordcount.txt.

```python
fp=open('article.txt','r')
text=fp.read()
fp.close()

words = text.split()
unique_words = set(words)

print(f"Uniques words {unique_words}")

freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

fp=open('wordCount.txt','w')
for word, count in freq.items():
    # print(f"{word} : {count}",end=" ")
    fp.write(f"{word} : {count}\n")
fp.close()
print("Word frequency analysis completed. Check wordcount.txt")
```
