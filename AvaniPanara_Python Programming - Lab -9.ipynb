### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def primeNumbers(n):
    for i in range(2,n//2+1):
        if n%i==0:
            return 0
            break;
    else:
        return 1;
```

```python
primeNumbers(22)
```

## 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
def primeNumberRange(n1,n2):
    li=[]
    for i in range(n1,n2):
        for j in range(2,i//2+1):
            if i%j==0:
                break;
        else:
            li.append(i);
    return li
```

```python
primeNumberRange(1,15)
```

### 05) Write a function that returns True if the given string is Palindrome or False otherwise.

```python
def palindromString(s):
    s=s.lower()
    li=s.split(" ")
    rev=li[::-1]
    print(li,rev)
    if rev==li:
        return True
    else:
        return False
```

```python
palindromString("Nayan")
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
def Fibonacci(n):
    if n<=1:
        return 1
    else:
        return Fibonacci(n-2)+Fibonacci(n-1)
```

```python
Fibonacci(10)
```

### 09) Write a function to get the name of the student based on the given rollno.
#### Example: Given dict1 = {101:'Ajay', 102:'Rahul', 103:'Jay', 104:'Pooja'} find name of student whose rollno = 103

```python
def getStudent(rollno,d1):
    for i in d1:
       if i==rollno:
           return d1[i]
       elif i not in d1.keys():
           return "Student not found"
```

```python
def DictNameAccess(d1,rollno):
    for i in range(len(d1)):
        if rollno==d1.keys():
            print(d1.keys()," ",d1.values())
    return d1.values()
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def invertDict(d1):
    d2={value:key for key,value in d1.items()}
    return d2
```

```python
size=int(input("Enter size of dict"))
d1={}
for i in range(size):
    key=input("Enter key")
    value=input("Enter value")
    d1[key]=value
print(f"Invert Dictionary: {invertDict(d1)}")
```

### 12) Write a function to check whether the given string is Pangram or not.
#### hint: Pangram is a string containing all the characters a-z atlest once.
#### "the quick brown fox jumps over the lazy dog" is a Pangram string.

```python
def pangramStringUsingFlag(s):
    s=s.lower()
    flag=False
    for i in s:
        if i in 'abcdefghijklmnopqrstuvwxyz':
            continue
        else:
            flag=False
    if flag==True:
        print("Pangram")
    else:
        print("Not")
```

```python
def pangramStringUserForElse(s):
    s=s.lower()
    s=s.replace(" ","")
    alphbets='abcdefghijklmnopqrstuvwxyz'
    for i in s:
        if i not in alphbets:
            print(f"'{s}' is NOT a Pangram.")
            break
    else:
        print(f"'{s}' IS a Pangram.")
```

```python
s=input("Enter string")
pangramStringUserForElse(s)
```

### 15) For the given list of names of students, extract the names having more that 7 characters. Use filter().

```python
def getNamesOfStudent(li):
    ans=list(filter(lambda i:len(i)>7 ,li))
    return ans
```

```python
li_size=int(input("Enter no of students"))
li=[]
for i in range(li_size):
    li.append(input("Enter student's name :"))
print(li)
getNamesOfStudent(li)
```

### 16) For the given list of names of students, convert the first letter of all the names into uppercase. use map().

```python
def firstCharUpper(li):
    ans=list(map(lambda i:i[0].upper()+i[1:],li))
    return ans
```

```python
li_size=int(input("Enter no of students"))
li=[]
for i in range(li_size):
    li.append(input("Enter student's name :"))
print(li)
firstCharUpper(li)
```

###  Positional Arguments

```python
def positionalArgument(num1,num2):
    return num1+num2
```

### Keyword Arguments

```python
def keywordArgument(num1,num2):
    return num1+num2
```

```python
keywordArgument(num1=23,num2=10)
```

### Default Arguments

```python
def defaultArgument(num1=0,num2=10):
    return num1+num2
```

```python
defaultArgument(12)# default value of b is 10 
```

### Variable Legngth Positional(*args)

```python
def sum_lengthPositional(*numbers):
        total = 0
        for num in numbers:
            total += num
        print(f"Sum: {total}")
```

```python
sum_lengthPositional(1, 2, 3)
sum_lengthPositional(10, 20, 30, 40, 50)
```

### variable length Keyword Arguments 

```python
def display_info(**details):
    for key, value in details.items():
        print(f"{key}: {value}")
```

```python
display_info(name="Eve", age=30, city="New York")
print()
display_info(name="Eve", age=30, city="New York",salary=10000)
```
