### 04) WAP to demonstrate the use of Pie chart.

```python
x = range(1,11,1)
cxMarks= [82,93,69,35,56,77,43,100,29,66]
cyMarks= [58,83,90,65,30,29,49,87,85,90]
colors = ['red','purple','green','blue','cyan','orange','pink','yellow','gray','brown']
plt.pie(x,labels=cxMarks,autopct='%1.2f%%',colors=colors,shadow=True)
plt.title('Pie Chart with Marks')
plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

```python
month= ['jan','feb','march','april','may','june','july','aug','sep','oct','nov','dec']
visiter=[12220,20000,30000,45000,80000,20905,23400,45350,67700,8000,39240,28170]
colors = ['red','purple','green','blue','cyan','orange','pink','yellow','gray','brown']
bars=plt.bar(month,visiter,width=0.3,color=colors)
for i in bars:
    plt.text(i.get_x()-0.2,i.get_height()+1000,f"{i.get_height()}")
plt.title('Bar Chart with visiter per month')
plt.show()
```
